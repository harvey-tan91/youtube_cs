# How to work with CSV Files

```python
import csv
import os

os.chdir(r'C:\Users\tanzh\Documents\Python\Other Materials')
```

```python
with open(r'C:\Users\tanzh\Documents\Python\Pandas for Everyone\datasets\tesla_stock_yahoo.csv','r') as csv_file: # this is the context manager line, open the file in read mode 
    csv_reader = csv.reader(csv_file) # we need to create a csv reader object with the reader method 
    
    print(csv_reader) # this is an iterable

    for line in csv_reader:
        print(line)
```

```python
with open(r'C:\Users\tanzh\Documents\Python\Pandas for Everyone\datasets\tesla_stock_yahoo.csv','r') as csv_file: # we open this file to read this file
    csv_reader = csv.reader(csv_file)
    with open('new_file_tabs.csv','w', newline='') as new_file: # we create a new file to write into
        csv_writer = csv.writer(new_file, delimiter='\t')
        for line in csv_reader:
            csv_writer.writerow(line)
```

```python
with open('new_file_tabs.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter='\t') # if our delimiter is a tab, we have to specify it in the agruement

    for line in csv_reader:
        print(line)
```

```python
with open(r'C:\Users\tanzh\Documents\Python\Pandas for Everyone\datasets\tesla_stock_yahoo.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    for line in csv_reader:
        print(line)
```
